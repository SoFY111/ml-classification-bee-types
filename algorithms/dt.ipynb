```python
import pandas as pd 
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import numpy as np

from sklearn import metrics
```

```python
data=pd.read_csv("../audioData.csv",encoding="ISO-8859-1")
col_name=['features', 'class']
data.columns=col_name
data.head()
```

```python
X=data['features']
y=data['class']
print(X)
print(y)
```

```python
import re
newX = []
for j in range(0, len(data['features'])):
    #print(j, data['feature'][j], )
    newP = []
    for i in data['features'][j].replace('\n', '').split(' '):
        if i != '':
            try:
                #print(float( re.sub("[^\d\.]", "", i)))
                newP.append(float( re.sub("[^\d\.]", "", i)))
            except:
                print(j, 'isnt ok.')
                continue
    
    print(j, 'is ok.')
    newX.append(newP)
```

```python
newX = np.array(newX)
newX.shape
```

```python
X_train, X_test, y_train,y_test= train_test_split(newX,y, test_size=0.2, random_state=1)
print(X_test)
```

```python
dt=DecisionTreeClassifier()
dt=dt.fit(X_train,y_train)
```

```python
y_pred=dt.predict(X_test)
print("Acc",metrics.accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
accuracy=accuracy_score(y_test, y_pred)
precision=precision_score(y_test, y_pred, average='macro')
recall=recall_score(y_test, y_pred, average='macro')
fmeasure=f1_score(y_test, y_pred, average='macro')
```

```python
print(accuracy)
print(precision)
print(recall)
print(fmeasure)
```
